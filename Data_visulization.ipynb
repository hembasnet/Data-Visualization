```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

import pandas as pd

#load data set
data = pd.read_csv('/content/Live (2).csv')
display(data.head(10))
```

```python
style.use("ggplot")

# Convert 'status_published' to datetime objects
data['status_published'] = pd.to_datetime(data['status_published'])

# Group by date and status type and count occurrences
status_counts = data.groupby([data['status_published'].dt.date, 'status_type']).size().unstack(fill_value=0)

# Plot the counts for each status type over time
status_counts.plot(figsize=(12, 6))

plt.xlabel('Date of Publication')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Status Type Over Time')
plt.legend(title='Status Type')
plt.tight_layout()
plt.show()
```

Let's visualize the distribution of 'num_likes' using a histogram.

```python
style.use("ggplot")

plt.figure(figsize=(10, 6))
plt.hist(data['num_likes'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Likes')
plt.tight_layout()
plt.show()
```

```python
style.use("ggplot")

reaction_columns = ['num_reactions', 'num_comments', 'num_shares', 'num_likes', 'num_loves', 'num_hahas', 'num_sads', 'num_angrys']

plt.figure(figsize=(15, 10)) # Adjust figure size for multiple plots

for i, col in enumerate(reaction_columns):
    plt.subplot(3, 3, i + 1) # Create subplots
    plt.hist(data[col], bins=50, color='skyblue', edgecolor='black')
    plt.xlabel(col.replace('num_', '').capitalize()) # Set appropriate x-label
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col.replace("num_", "").replace("_", " ").capitalize()}')

plt.tight_layout()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(10, 6))
data['status_type'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.xlabel('Status type')
plt.ylabel('Frequency')
plt.title('Distribution of Status Type')
plt.tight_layout()
plt.show()
```
